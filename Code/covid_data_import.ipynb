```python
import pandas as pd
```

```python
# read data from csv file
covid_df = pd.read_csv("../Resources/owid-covid-data.csv")
covid_df.head()
```

```python
# save the necessary columns into new dataframe
covid_df1=covid_df[["continent","location","date","total_cases","new_cases"]]
covid_df1.head()
```

```python
# get the data for Canada and United States
covid_df2=covid_df1[(covid_df1['location']=='Canada') | (covid_df1['location']=='United States')]               
covid_df2
```

```python
# get the data from 2020-01-01 to 2020-07-31
covid_df3 = covid_df2[(covid_df2['date']>='2020-01-01') & (covid_df2['date']<='2020-08-12')]   
covid_df3
```

```python
# reorganizing the columns
covid_df4 = covid_df3[["continent","location","date","new_cases","total_cases"]]
covid_df4
```

```python
# rename the columns
covid_df5= covid_df4.rename(columns={"continent":"Continent","location":"Country", "date":"Date", \
                                     "new_cases":"Daily New Cases", "total_cases":"Total Cases"})
covid_df5
```

```python
# reset the index and get the final cleaned covid case data
covid_data = covid_df5.reset_index(drop=True)
covid_data
```
