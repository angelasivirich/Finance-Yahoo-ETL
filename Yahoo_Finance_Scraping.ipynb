```python
def last_day_of_month_in_timestamp(month):
    next_month = datetime.date(2020, month, 1).replace(day=28) + datetime.timedelta(days=4)
    date = (next_month - datetime.timedelta(days=next_month.day))
    element = datetime.datetime.strptime(str(date),"%Y-%m-%d")
    return int(datetime.datetime.timestamp(element))

def first_day_of_the_month_in_timestamp(month):
    date = datetime.datetime(2020, month, 1)
    element = datetime.datetime.strptime(str(date),"%Y-%m-%d %H:%M:%S")
    return int(datetime.datetime.timestamp(element))
```

```python

def stock_dataframes(stock):
    table_list = []
    for month in range(1, 9):
        start_date = first_day_of_the_month_in_timestamp(month)
        end_date = last_day_of_month_in_timestamp(month)

        #crude_oil = f'https://finance.yahoo.com/quote/CL%3DF/history?period1={start_date}&period2={end_date}&interval=1d&filter=history&frequency=1d'
        stock_data = f'https://finance.yahoo.com/quote/{stock}/history?period1={start_date}&period2={end_date}&interval=1d&filter=history&frequency=1d'

        tables = pd.read_html(stock_data)
        table_list.append(tables)
        
    return table_list
```

```python
Pton= stock_dataframes("PTON")

jan_Pton_df = Pton[0][0]
feb_Pton_df = Pton[1][0]
mar_Pton_df = Pton[2][0]
apr_Pton_df = Pton[3][0]
may_Pton_df = Pton[4][0]
jun_Pton_df = Pton[5][0]
jul_Pton_df = Pton[6][0]
aug_Pton_df = Pton[7][0]
```

```python

Pton_df = [aug_Pton_df, jul_Pton_df, jun_Pton_df, may_Pton_df, apr_Pton_df, mar_Pton_df, feb_Pton_df, jan_Pton_df]
pton_df = pd.concat(Pton_df)
pton_df
```

```python
pton_df['Date']=pd.to_datetime(pton_df['Date'], errors='coerce').copy()
```

```python
pton_df.reset_index(inplace=True)
pton_df.drop(columns='index', inplace=True)
pton_df=pton_df[pton_df.Open!='*Close price adjusted for splits.**Adjusted close price adjusted for both dividends and splits.']
```

```python
PTON_df=pton_df[['Date','Open','Close*']].copy()
PTON_df.columns=['Date','Pton_open','Pton_close']
PTON_df['Pton_open']=PTON_df['Pton_open'].astype('str', copy=True).astype('float', copy=True)
PTON_df['Pton_close']=PTON_df['Pton_close'].astype('str', copy=True).astype('float', copy=True)

```

```python
PTON_df['Pton_pert_change']=round(((PTON_df['Pton_close']-PTON_df['Pton_open'])/(PTON_df['Pton_open']))*100,2)
PTON_df.head()
```
