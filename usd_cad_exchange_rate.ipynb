```python
#Dependencies
import requests
from pprint import pprint
import json
import pandas as pd 
import matplotlib
```

```python
#Establish base url 
base_url = "https://www.bankofcanada.ca/valet"
```

```python
#requires: seriesNames, groupName, or listName
#format and query parameters optional
```

```python
query = "/observations/FXUSDCAD/json"
```

```python
url = base_url + query
```

```python
response = requests.get(url)
```

```python
response_json = response.json()
```

```python
print(json.dumps(response_json, indent=4, sort_keys = True))
```

```python
#JSON in DataFrame
exchange_rate_df = pd.DataFrame(response_json['observations'])
```

```python
exchange_rate_df.head(1000)
```

```python
#Renaming the column headers
renamed_df = exchange_rate_df.rename(columns={"d": "Date"})
```

```python
renamed_df.head(1000)
```

```python
renamed_df = renamed_df.drop(renamed_df[renamed_df.Date < '2020-01-01'].index)
renamed_df
#Tested it with 2019-12-31 and no result for 2020-01-01
```

```python
#Set Date as Index 
reindexed_df = renamed_df.set_index('Date')
```
